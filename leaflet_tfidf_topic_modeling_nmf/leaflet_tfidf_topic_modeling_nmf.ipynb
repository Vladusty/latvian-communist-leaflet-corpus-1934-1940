```python
# Lightweight Topic Modeling for Latvian Leaflets (ZIP Upload)
# Uses only scikit-learn + pandas + matplotlib.
# Method: TF–IDF → NMF topics (alternatively LDA or BERTopic).
# ZIP of .txt files required.

# %% [markdown]
# ## How to use
# 1) Run cell 1: Upload ZIP with your `.txt` leaflets.
# 2) Run cells 2→end. You will get: topic-word lists, doc→topic table, charts, CSV export.
# 3) You can change `N_TOPICS` and rerun the modeling cell.

# %%
# 1) Upload ZIP and extract TXT files
import os, re, zipfile, io
import pandas as pd
from google.colab import files
from IPython.display import display

print("Please upload your ZIP file (with .txt leaflets inside):")
uploaded = files.upload()
if not uploaded:
    raise SystemExit("No file uploaded.")

zip_name = next(iter(uploaded.keys()))
if not zip_name.endswith('.zip'):
    raise SystemExit("Please upload a .zip archive.")

extract_dir = "leaflets_zip"
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(io.BytesIO(uploaded[zip_name]), 'r') as zf:
    zf.extractall(extract_dir)

files_list = [os.path.join(extract_dir, f) for f in os.listdir(extract_dir) if f.lower().endswith('.txt')]
print(f"Extracted {len(files_list)} .txt files → {extract_dir}")
if not files_list:
    raise SystemExit("ZIP contained no .txt files.")

# %%
# 2) Parse corpus and minimal Latvian cleaning
LATVIAN_STOPWORDS = set('''un par lai kas bet arī kā tas tādas tādi tādas tās šis šie šīs bija būs bez pie pret vai kad tad tik ar no uz mēs jūs viņi viņas viņš viņa kuri kuras kuru kādu kādas kādi kāds viss visi visas jo vai nu gan drīz vien vēl jau pat kāpēc kurp ne nav tiek tiekam tikt tika vairāk mazāk ļoti tikai nekā taču '''.split())
META_STOP = {"id","file_name","title","author","date","print_run","typography_name","source","text","revl"}
STOPWORDS = LATVIAN_STOPWORDS | META_STOP

meta_re = re.compile(r"\ntext:\s*", flags=re.IGNORECASE)
meta_re2 = re.compile(r"\ntext\n", flags=re.IGNORECASE)

rows = []
for path in files_list:
    with open(path, 'r', encoding='utf-8', errors='replace') as fh:
        raw = fh.read()
    # extract body after 'text:' marker if present
    parts = meta_re.split(raw, maxsplit=1)
    if len(parts)==2:
        body = parts[1]
    else:
        parts2 = meta_re2.split(raw, maxsplit=1)
        body = parts2[1] if len(parts2)==2 else raw
    # clean
    t = body.lower()
    t = re.sub(r"\b\d{2,}\b", " ", t)
    t = re.sub(r"[^\w\s\-\ā\č\ē\ģ\ī\ķ\ļ\ņ\ŗ\š\ū\ž]", " ", t)
    t = re.sub(r"\s+", " ", t).strip()
    tokens = [w for w in t.split() if len(w)>2 and w not in STOPWORDS]
    clean = " ".join(tokens)
    rows.append({"file_name": os.path.basename(path), "raw": raw, "text": body.strip(), "clean": clean})

df = pd.DataFrame(rows)
display(df.head(3))
print(f"Documents ready: {len(df)}")

# %%
# 3) TF–IDF vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

# You can tweak ngrams and min_df for stability
vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=3, max_df=0.95)
X = vectorizer.fit_transform(df['clean'].values)
feature_names = vectorizer.get_feature_names_out()
print(X.shape)

# %%
# 4) NMF topic modeling (change N_TOPICS to tune granularity)
from sklearn.decomposition import NMF

N_TOPICS = 6  # ← adjust if needed (e.g., 6–20)
N_TOP_WORDS = 15

nmf = NMF(n_components=N_TOPICS, init='nndsvda', random_state=42, max_iter=500)
W = nmf.fit_transform(X)  # doc-topic
H = nmf.components_       # topic-term

# Top words per topic
topics_words = []
for k in range(N_TOPICS):
    top_idx = H[k].argsort()[::-1][:N_TOP_WORDS]
    words = [feature_names[i] for i in top_idx]
    topics_words.append(words)
    print(f"\nTopic {k}: ", ", ".join(words))

# Assign dominant topic to documents
import numpy as np

df['topic'] = W.argmax(axis=1)
df['topic_score'] = W.max(axis=1)
df['topic_keywords'] = df['topic'].apply(lambda k: ", ".join(topics_words[int(k)]))

display(df[['file_name','topic','topic_score','topic_keywords']].head(10))


# --- Semantic labels for topics by keywords ---
import re

def label_topic(words):
    txt = " ".join(words)
    if re.search(r'spānij|spāņu', txt):                   return 'Spain'
    if re.search(r'\bsieviet', txt):                      return 'Women'
    if re.search(r'sarkan|palīdz|cietum|politiesl', txt): return 'Aid(SP)'   # Sarkanā Palīdzība
    if re.search(r'jaunatn|skol', txt):                   return 'Youth'
    if re.search(r'sprs|padomju|kar|vāci|miers|baltij', txt): return 'Foreign/War'
    if re.search(r'strādniek|alga|arodbiedr|bezdarb', txt):   return 'Labour'
    return 'Other'

topic_labels = {k: label_topic(topics_words[k]) for k in range(N_TOPICS)}
df['topic_label'] = df['topic'].map(topic_labels)

display(pd.DataFrame([
    {'topic_id': k, 'label': topic_labels[k], 'top_words': ", ".join(topics_words[k][:6])}
    for k in range(N_TOPICS)
]))


# %%
# 5) Quick charts (topic sizes & strengths) — matplotlib only
import matplotlib.pyplot as plt

sizes = df.groupby('topic').size().reset_index(name='count').sort_values('topic')
plt.figure()
plt.bar(sizes['topic'], sizes['count'])
plt.title('Topic sizes (NMF)')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.show()

plt.figure()
plt.hist(df['topic_score'], bins=30)
plt.title('Document topic dominance (max W)')
plt.xlabel('Max topic weight per document')
plt.ylabel('Frequency')
plt.show()


# %%
# 6) 2D scatter — colors=topics, markers=years (NO axis flipping)
from sklearn.decomposition import TruncatedSVD
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

svd = TruncatedSVD(n_components=2, random_state=42)
coords = svd.fit_transform(X)

rng = np.random.default_rng(42)
jitter = 0.0025
x = coords[:, 0] + rng.normal(0, jitter, len(coords))
y = coords[:, 1] + rng.normal(0, jitter, len(coords))

label_palette = {
    'Labour':      '#1f77b4',
    'Aid(SP)':     '#2ca02c',
    'Youth':       '#9467bd',
    'Women':       '#d62728',
    'Foreign/War': '#ff7f0e',
    'Spain':       '#8c564b',
    'Other':       '#7f7f7f',
}

def extract_year(name):
    m = re.search(r'19(3[4-6])', name)
    return int(m.group(0)) if m else None
if 'year' not in df.columns:
    df['year'] = df['file_name'].apply(extract_year)

year_markers = {1934: 'o', 1935: '^', 1936: 's', None: 'x'}

labels = df['topic_label']
years  = df['year']

fig, ax = plt.subplots(figsize=(10,7))

label_counts = labels.value_counts()
for lbl in sorted(label_counts.index):
    mask_lbl = (labels == lbl).values
    for yr, marker in year_markers.items():
        mask = mask_lbl & (years == yr)
        if not mask.any():
            continue
        ax.scatter(
            x[mask], y[mask],
            c=label_palette.get(lbl, '#7f7f7f'),
            s=70,
            marker=marker,
            edgecolors='k', linewidths=0.3,
            label=f"{lbl} {yr}" if yr else f"{lbl}"
        )

show_year_centers = True
if show_year_centers:
    year_order = [1934, 1935, 1936]
    centers = {}
    for yr in year_order:
        m = (df['year'] == yr).values
        if m.sum() > 0:
            cx, cy = np.mean(x[m]), np.mean(y[m])
            centers[yr] = (cx, cy)
            ax.scatter([cx], [cy], s=220, facecolors='none', edgecolors='k', linewidths=2, zorder=20)
            ax.text(cx, cy+0.015, str(yr), ha='center', va='bottom',
                    fontsize=11, weight='bold', zorder=21)
    for i in range(len(year_order)-1):
        y1, y2 = year_order[i], year_order[i+1]
        if y1 in centers and y2 in centers:
            (x1, y1c), (x2, y2c) = centers[y1], centers[y2]
            ax.plot([x1, x2], [y1c, y2c], linestyle='--', color='k', linewidth=1.2, zorder=19)
            ax.annotate("", xy=(x2, y2c), xytext=(x1, y1c),
                        arrowprops=dict(arrowstyle="->", color='k', lw=1.6),
                        zorder=19)

# 6) legends
LEG_FONTSIZE = 9

plt.subplots_adjust(right=0.80)
side_ax = fig.add_axes([0.82, 0.10, 0.16, 0.80])  # x, y, w, h в долях фигуры
side_ax.axis('off')

leg_topics = side_ax.legend(
    handles=topic_handles, title="Topics",
    loc='upper left', bbox_to_anchor=(0.0, 1.0),
    frameon=True, fontsize=LEG_FONTSIZE, title_fontsize=LEG_FONTSIZE+1
)
side_ax.add_artist(leg_topics)

_ = side_ax.legend(
    handles=year_handles, title="Year",
    loc='upper left', bbox_to_anchor=(0.0, 0.35),
    frameon=True, fontsize=LEG_FONTSIZE, title_fontsize=LEG_FONTSIZE+1
)




# Topic mix by year
tab = (df.dropna(subset=['year'])
         .pivot_table(index='year', columns='topic_label', values='file_name', aggfunc='count', fill_value=0))
tab_pct = tab.div(tab.sum(axis=1), axis=0).round(3)
display(tab)       # absolute
display(tab_pct)   # relative

# --- stacked bar (robust: fill missing topic columns with 0 and keep a fixed order)
import matplotlib.pyplot as plt

colors = {
    'Labour':'#1f77b4','Aid(SP)':'#2ca02c','Youth':'#9467bd',
    'Women':'#d62728','Foreign/War':'#ff7f0e','Spain':'#8c564b','Other':'#7f7f7f'
}

tab_pct_plot = tab_pct.reindex(columns=list(colors.keys()), fill_value=0)

ax = tab_pct_plot.plot(
    kind='bar', stacked=True,
    color=[colors[c] for c in tab_pct_plot.columns],
    figsize=(9,6)
)
ax.set_title('Topic shares by year')
ax.set_ylabel('Share')
ax.set_xlabel('Year')
ax.set_ylim(0, 1)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()

# --- 1935 vs 1934: what words have risen and fallen ---
import numpy as np, pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from IPython.display import display

# берём только документы 1934 и 1935
sub35 = df[df['year'].isin([1934, 1935])].copy()
if sub35.empty or sub35['year'].nunique() < 2:
    print("Not enough dates to compare 1934 and 1935")
else:
    vec35 = TfidfVectorizer(ngram_range=(1,2), min_df=3, max_df=0.95)
    X35 = vec35.fit_transform(sub35['clean'])
    vocab35 = np.array(vec35.get_feature_names_out())

    m1934 = X35[(sub35['year'].values == 1934)].mean(axis=0).A1
    m1935 = X35[(sub35['year'].values == 1935)].mean(axis=0).A1

    diff = m1935 - m1934
    chg35 = pd.DataFrame({
        'term': vocab35,
        '1934': m1934,
        '1935': m1935,
        'change': diff
    }).sort_values('change', ascending=False)

    print("↑ Increased in 1935 (top-20):")
    display(chg35.head(20))
    print("↓ Decreased to 1935 (top-20):")
    display(chg35.tail(20))



from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd


mask = df['year'].isin([1934, 1935, 1936]).values
X2 = np.column_stack([coords[mask, 0], coords[mask, 1]])
y2 = df.loc[mask, 'year'].astype('int64').values

clf = LogisticRegression(max_iter=200)  # default: lbfgs + multinomial when 2 classes
acc = cross_val_score(clf, X2, y2, cv=5).mean()

y2_off = (y2 - 1934).astype('int64')
baseline = np.bincount(y2_off).max() / y2_off.size

print(f"Accuracy (coords→year) = {acc:.2f} vs baseline {baseline:.2f}")



import numpy as np, pandas as pd

df['yr_bin'] = np.where(df['year']==1936, '1936', np.where(df['year'].isin([1934,1935]), '1934-35', None))
sub = df[df['yr_bin'].notna()].copy()

from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(ngram_range=(1,2), min_df=3, max_df=0.95)
Xsub = vec.fit_transform(sub['clean'])
vocab = np.array(vec.get_feature_names_out())

m1 = Xsub[sub['yr_bin'].values=='1934-35'].mean(axis=0).A1
m2 = Xsub[sub['yr_bin'].values=='1936'].mean(axis=0).A1
chg = pd.DataFrame({'term':vocab, 'early':m1, 'late':m2, 'change':m2-m1}).sort_values('change', ascending=False)

print("↑ Increased in 1936 (top-20):")
display(chg.head(20))
print("↓ Decreased to 1936 (top-20):")
display(chg.tail(20))


# %%
# 7) Export CSV
out = df[['file_name','topic','topic_score','topic_keywords']]
CSV_NAME = 'leaflets_topics_nmf.csv'
out.to_csv(CSV_NAME, index=False)
from google.colab import files as colab_files
colab_files.download(CSV_NAME)
print(f"Saved {CSV_NAME}")
```
